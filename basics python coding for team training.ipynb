```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# how to create numeric variables, scatter plot and label the axis
x=[1,2,3,4,5,6,7,8]
y=[2,4,6,8,10,12,14,16]
plt.bar(x,y)
plt.xlabel("# of days")
plt.ylabel("# of exercise")
plt.show()

```

```python
#how to dataframe variables, plot and add axis labels
df = pd.DataFrame({'x':[1,2,3,4,5,6,7,8],
'y':[2,4,6,8,10,12,14,16]})

df.plot(kind='bar',x='x',y='y',color='green')
plt.xlabel("# of days")
plt.ylabel("# of exercise")
plt.show()
```

```python


#label it
y=[1,2,3]
x=['a','b','c']

plt.bar(x,y)
plt.xlabel("# of days")
plt.ylabel("# of exercise")

for i in range(len(x)):
    plt.text(i, y[i], y[i], ha = 'center')
plt.show()
```
